```python
# Import dependencies
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import mplcursors
```

### first exploration - Analysis: Are there song characteristic preferences that change over time?

```python
#import csv fiiles and safe as Data Frames
spotify_cln_df = pd.read_csv('Clean/cleaned_spotify_data.csv')
genre_country_cln_df = pd.read_csv('Clean/country_genres_data.csv')

spotify_cln_df.head()
```

```python
#grouping data by year and taking mean of characteristics
spotify_group = spotify_cln_df.groupby('Year').mean()
spotify_group
```

```python
# create a line chart to examine all characteristics over time
#establish values to plot
columns_to_plot = ["Bpm", "Energy", "Danceability", "Liveness", "Valence", "Duration (in seconds)", "Acousticness", "Speechiness", "Spotify Rating"]

#create plot
spotify_group[columns_to_plot].plot()
    
#add legend
plt.legend(loc='upper right', bbox_to_anchor=(1.5, 1))

#add labels and title
plt.xlabel("Year")
plt.ylabel("Value")
plt.title("Trends over time")

#display plot
plt.savefig('Analyses_and_outputs/Trends_Over_Time.jpg', dpi=300, bbox_inches='tight')
plt.show
```

```python
#saving important values for use in markdown
long_duration = round(spotify_group.loc[spotify_group.index == 1999, 'Duration (in seconds)'], 1)
long_duration
short_d = round(spotify_group.loc[spotify_group.index == 2023, 'Duration (in seconds)'], 1)
short_d
```

```python
# Establish values to plot
columns_to_plot = ["Bpm", "Energy", "Danceability", "dB", "Liveness", "Valence", "Duration (in seconds)", "Acousticness", "Speechiness", "Spotify Rating"]
y_vals = spotify_group[columns_to_plot]
x_vals = spotify_group.index  # Use the index as x-axis values

# Determine the number of rows and columns for subplots
num_plots = len(columns_to_plot)
num_rows = int(num_plots / 2)  # Assuming 2 columns per row
num_cols = 2  # 3 columns per row

#  Creating the figure and subplots based on the number of rows and columns.
fig, axes = plt.subplots(num_rows, num_cols, figsize=(25, 20))
fig.tight_layout(pad=3.0)  # Adjust the spacing between subplots

# Iterate over columns and create individual plots
for i, column in enumerate(columns_to_plot):
    row = i // num_cols
    col = i % num_cols
    ax = axes[row, col] if num_rows > 1 else axes[col]  # Handle single-row case
    
    ax.plot(x_vals, y_vals[column])
    ax.axhline(y_vals[column].mean(), color='r', linestyle='--', label='Mean') #add mean line for each plot
    ax.set_title(column, fontweight='bold')
    ax.set_xlabel('Year')
    ax.set_ylabel('Value')
    
     # Adjust y-axis scale based on min and max values
    min_value = y_vals[column].min()
    max_value = y_vals[column].max()
    y_offset = 10  # Adjust the offset as needed
    ax.set_ylim(min_value - y_offset, max_value + y_offset)


# Display the plots
plt.savefig('Analyses_and_outputs/Spotify_Characteristics_v_Time.jpg', dpi=300, bbox_inches='tight')
plt.show()
```

### 1.
<p>In my opinion, the most eye-catching trend is between 1995 and 2000. If you look at the years 1996-1999, you can see a period of sharp decline in the following plots: <br>
    <ul>

  <li>Energy</li>
</ul>
<p>While you can see a distinct rise during that same time period in the following plots:
       <ul>
  <li>Speechiness</li>
  <li>Acousticness</li>
</ul>
<p>Our hypothesis is that during the late 1990's the public had a phase where slow, speechy, and acoustic-y songs were highly appreciated and quite popular.</p>
 
#### 2.
<p> The second trend we observed was how the duration of popular songs has slowly decreased over time. From 1999-2002 the average length of song in the top 100 was over 240s (4 minutes), folowing this is a stable period where song length remains between 220-240s (3.5-4 min). However, after 2016, things begin to change; by 2019 the mean song length is 200s and by 2023 it drops down to 192s which equates to 3min, 9s per song.<br>
<p> At the peak, songs were averaging 250s and by 2023 they have dropped to ~190s. This is close to a full minute shorter. Our hypothesis is that as everyday technology consumption has grown exponentially since 2000, along with the shortening of attention-spans and increase in screen time, musicians have been encouraged to produce shorter and shorter tracks. There is a lot of competition for attention these days and it could be considered risky to create longer trakcs as it is more time for a listener to skip or move on to the next thing.
