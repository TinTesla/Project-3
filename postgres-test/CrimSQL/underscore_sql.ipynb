```python
import pandas as pd
```

```python
 # The path to our CSV file
file = "countries.csv"

# Read our Crowdfunding data into pandas
df = pd.read_csv(file, sep=",")
df.head()
```

```python
df["ID"]=df["Country"]
```

```python
df["ID"]=df["ID"].apply(lambda x:x.lower().replace(" ","_"))
```

```python
df["ID"].head()
```

```python
df.head()
```

```python
df["Country"]=df["ID"]
df.head()
```

```python
df = df.drop(columns="ID")
df.head()
```

```python
df.to_csv('under_clean_countries.csv',index=False)
```
